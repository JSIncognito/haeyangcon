```python
import pandas as pd
import os
import re
```

```python
os.getcwd()
dir_path = "../../../01.Data/13.정선해양관측/"
```

```python
# 해당 위치의 모든 파일명을 list 형태로 가져옴 
fd_filenm = os.listdir(dir_path)
# fd_filenm
```

```python
# 확장자가 csv인 경우에만 리스트에 담도록 하는 구문
fd_filenm=[filenm for filenm in fd_filenm if filenm[-3:]=='csv']
# fd_filenm
```

```python
# 파일목록 구분 정규표현식 
cmc_cp = re.compile(r"(\w+(chemical)\w.+)")
mtr_cp = re.compile(r"(\w+(meteorological)\w.+)")
# mt = r"(\w+(chemical)\w.+)"
cmc_filelist = []
mtr_filelist = []
for nm in fd_filenm:
    cmc_m = cmc_cp.match(nm)
    mtr_m = mtr_cp.match(nm)
    if cmc_m != None :
        # print("cmc", "=="*10)
        cmc_filelist.append(cmc_m.group())
    elif mtr_m != None : 
        # print("mtr", "=="*10)
        mtr_filelist.append(mtr_m.group())
```

```python

# chemical 파일 불러오기/합치기
def cmcConcatReadCsv(arg_fnm_list):
    global dir_path
    fd_path = dir_path
    # chemical 컬럼
    df_cmc_col = ["project_name", "station", "yyyy-mm-dd hh:mm:ss", "Longitude [degrees_east]"
                , "Latitude [degrees_north]", "bot.depth[m]", "수심(m)", "dep_flag", "염분[psu]", "sal_qc"
                , "수온[℃]", "wtrtmp_qc", "용존산소[mg/L]", "disoxy_qc", "규산염[μM]", "reasil_qc", "아질산성질소[μM]"
                , "nitnit_qc", "인산염인[μM]", "po4p_qc", "질산성질소[μM]", "natnit_qc", "투명도[m]", "tr_qc"]
    err_li=[]
    dfs = pd.DataFrame()

    cnt = 0
    for fnm in arg_fnm_list:
        print(fnm)
        try:
            df = pd.read_csv(fd_path+fnm, encoding='cp949', skiprows=26)
            print("read csv ", "="*50)
            print(df.shape)
            cnt += 1
        # # 오류가 발생된 경우 
        except:
            print("오류발생:", fnm)
            err_li.append(fnm)
        # # 정상적으로 read_csv 가 수행된 경우 
        else:
        #     # 파일을 합치는 중에 컬럼 열은 6열, 데이터는 7열인 파일이 있어서 변경하기 위해 반영
        #     if df.shape[1] == 7:
        #         df = df.iloc[:, :-1]
            print("else", "="*50)
            # df.columns = df_mtr_col
            print(df.columns)
            print("else", "="*50)
            dfs = pd.concat([dfs, df])

    print(err_li)
    print("", cnt)
    return dfs
```

```python
# meteorological 파일 불러오기/합치기
def mtrConcatReadCsv(arg_fnm_list):
    global dir_path
    fd_path = dir_path

    # meteorological 컬럼
    df_mtr_col = ["project_name", "station", "yyyy-mm-dd hh:mm:ss", "Longitude [degrees_east]"
                , "Latitude [degrees_north]", "bot.depth[m]", "수심(m)", "dep_flag", "기압[hPa]", "airpre_qc"]
    err_li=[]
    dfs = pd.DataFrame()

    cnt = 0
    for fnm in arg_fnm_list:
        print(fnm)
        
        try:
            df = pd.read_csv(fd_path+fnm, encoding='cp949', skiprows=26)
            print("read csv ", "="*50)
            print(df.shape)
            cnt += 1
        # 오류가 발생된 경우 
        except:
            print("오류발생:", fnm)
            err_li.append(fnm)
        # 정상적으로 read_csv 가 수행된 경우 
        else:
        #     # 파일을 합치는 중에 컬럼 열은 6열, 데이터는 7열인 파일이 있어서 변경하기 위해 반영
        #     if df.shape[1] == 7:
        #         df = df.iloc[:, :-1]
            print("else", "="*50)
            # df.columns = df_mtr_col
            print(df.columns)
            print("else", "="*50)
            dfs = pd.concat([dfs, df])

    print(err_li)
    print("", cnt)
    return dfs    
```
